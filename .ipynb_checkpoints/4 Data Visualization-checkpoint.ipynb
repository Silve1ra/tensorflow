```python
x = np.arange(0,10)
```

```python
x
```

```python
y = x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'*')
```

```python
plt.plot(x,y,'red')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r--')
plt.xlim(0,4)
```

```python
plt.plot(x,y,'r--')
plt.xlim(0,4)
plt.ylim(0,10)
```

```python
plt.plot(x,y,'r--')
plt.xlim(0,4)
plt.ylim(0,10)
plt.title("GRAOS")
plt.xlabel("COR")
plt.ylabel("TAMANHO")
```
